```python
import rvtx
import numpy as np
from IPython.display import Image
```

```python
try:
    if RVTX_WINDOW_CREATED:
        print("rvtx window already created.")
except:
    window = rvtx.Window(width=1280, height=720, title="Ravioli Viewer", shown=False)
    RVTX_WINDOW_CREATED = True

engine = rvtx.Engine(window)

def screenshot(width = 640, height = 360, scene = rvtx.main_scene):
    display(Image(data=engine.screenshot(scene), width=width, height=height))
```

```python
molecule = rvtx.load_molecule("./data/1AGA.mmtf")
```

```python
camera = rvtx.create_camera(rvtx.Transform([0, 0, -20]), rvtx.Camera.Target(molecule.aabb))
rvtx.set_main_camera(camera)
engine.update()
```

```python
point_cloud = rvtx.create_point_cloud([molecule.aabb.centroid + [5, 0, 0]], molecule.aabb.radius - 5, [1, 0, 0])
screenshot()
```

```python
point_cloud = rvtx.create_point_cloud([molecule.aabb.centroid - [5, 0, 0]], molecule.aabb.radius - 5, [0, 1, 0])
screenshot() # should only show the point cloud in green
```

```python
del point_cloud
screenshot() # should not show any point cloud
```

```python
point_cloud = rvtx.main_scene.create_point_cloud([molecule.aabb.centroid - [5, 0, 0]], molecule.aabb.radius - 5, [0, 1, 0])
screenshot() # should only show the point cloud in green
```

```python
point_cloud = rvtx.main_scene.create_point_cloud([molecule.aabb.centroid + [5, 0, 0]], molecule.aabb.radius - 5, [1, 0, 0])
screenshot() # should show both red and green point cloud
```

```python
scene = rvtx.Scene()
camera = rvtx.create_camera(rvtx.Transform([0, 0, -20]), rvtx.Camera.Target(molecule.aabb))
scene.set_main_camera(camera)
engine.update(scene)
```

```python
screenshot(scene=scene)
```
